### 테크닉 이용이력을 집계하기

```python
#연월(202012)컬럼 추가
uselog['usedate'] = pd.to_datetime(uselog['usedate'])
uselog['연월'] = uselog['usedate'].dt.strftime('%Y%m')

#연월, customer_id 기준으로 groupby => 각 연월마다 어느 회원이 몇번 왔는지 파악 가능
uselog_month = uselog.groupby(['연월', 'customer_id']).count()
uselog_month.rename(columns={'log_id':'count'}, inplace=True) #원본에 엎어치기

#usedate 컬럼 drop
del uselog_month['usedate']

uselog_month
```

```python
#각종 통계량 구하기
uselog_month.groupby("customer_id").mean()
```

```python
uselog_month.groupby("customer_id").max()
```

```python
uselog_month.groupby("customer_id").min()
```

```python
#agg => 
uselog_month.groupby("customer_id").agg(["mean","max","min"])
```

```python
#dt.weekday
#0~6 : 월~일

df = pd.DataFrame({'Birth':['2019-01-01 09:10:00',
                            '2019-01-08 09:20:30',
                            '2019-02-01 10:20:00',
                            '2019-02-02 11:40:50',
                            '2019-02-28 15:10:20',
                            '2019-04-10 19:20:50',
                            '2019-06-30 21:20:50',
                            '2019-07-20 23:30:59']})

df['Birth'] = pd.to_datetime(df['Birth'])
df['Birth'] = df['Birth'].dt.weekday
df
```

```python
#uselog[usedate] 컬럼에 해당하는 요일은 몇요일인지 구하기 

#weekday 컬럼추가
uselog['weekday'] = uselog['usedate'].dt.weekday

#
uselog_weekday = uselog.groupby(['customer_id','연월','weekday'], as_index=False).count()#[['customer_id','연월','log_id']]

#rename
uselog_weekday.rename(columns={"log_id":"count"}, inplace=True)

#컬럼 drop
del uselog_weekday["usedate"]

uselog_weekday.head()
```

```python
uselog_weekday.groupby(['customer_id'], as_index=False).max()[["customer_id","count"]]
```
